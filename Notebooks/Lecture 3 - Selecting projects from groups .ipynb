```python
import pyomo.environ as pe
import pyomo.opt as po
```

```python
model = pe.ConcreteModel()
```

```python
model.P = pe.RangeSet(1,np)
model.Y = pe.RangeSet(1, ny)
model.G = pe.RangeSet(1, ng)
```

```python
model.x = pe.Var(model.P, domain = pe.Binary)
```

```python
#Objective 
objExpr = sum(model.x[p] * npv[p] for p in model.P)
model.obj = pe.Objective(expr = objExpr, sense = pe.maximize)
```

```python
## Two sets of constraints
model.cnstrY = pe.ConstraintList()
for y in model.Y:
    expression = sum(c[y,p] * model.x[p] for p in model.P) <= b
    model.cnstrY.add(expression)

model.cnstrG = pe.ConstraintList()
for j in model.G:
    expression = sum(model.x[p] for p in model.P if g[p] == j) <= 1
    model.cnstrG.add(expression)
```

```python
solver = po.SolverFactory('gurobi')
result = solver.solve(model, tee = True)
```

```python
print(result.solver.status)
print(result.solver.termination_condition)
```
